```python
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense

# -----------------------------
# Training Corpus
# -----------------------------
train_sentences = [
    "I love natural language processing",
    "I love machine learning",
    "language models are useful",
    "machine learning is powerful"
]

# -----------------------------
# Test Corpus
# -----------------------------
test_sentences = [
    "I love language models",
    "machine learning models"
]

# -----------------------------
# Tokenization
# -----------------------------
tokenizer = Tokenizer()
tokenizer.fit_on_texts(train_sentences)

vocab_size = len(tokenizer.word_index) + 1

train_sequences = tokenizer.texts_to_sequences(train_sentences)

# -----------------------------
# Create Training Data
# -----------------------------
X = []
y = []

for seq in train_sequences:
    for i in range(1, len(seq)):
        X.append(seq[:i])
        y.append(seq[i])

max_len = max(len(seq) for seq in X)

X = pad_sequences(X, maxlen=max_len)
y = np.array(y)

# -----------------------------
# Neural Language Model
# -----------------------------
model = Sequential()

model.add(Embedding(vocab_size, 10, input_length=max_len))
model.add(LSTM(64))
model.add(Dense(vocab_size, activation='softmax'))

model.compile(loss='sparse_categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

# -----------------------------
# Train Model
# -----------------------------
model.fit(X, y, epochs=200, verbose=0)

# -----------------------------
# Next Word Prediction Function
# -----------------------------
def predict_next_word(text):
    seq = tokenizer.texts_to_sequences([text])[0]
    seq = pad_sequences([seq], maxlen=max_len)

    pred = model.predict(seq, verbose=0)
    predicted_word = tokenizer.index_word[np.argmax(pred)]

    return predicted_word

# -----------------------------
# Testing
# -----------------------------
print("Next word after 'I love':", predict_next_word("I love"))
print("Next word after 'machine':", predict_next_word("machine"))
```

```python
import math
import torch
import torch.nn as nn
import torch.optim as optim
from collections import Counter

# -----------------------------
# Training Corpus
# -----------------------------
train_sentences = [
    "I love natural language processing",
    "I love machine learning",
    "language models are useful",
    "machine learning is powerful"
]

# -----------------------------
# Test Corpus
# -----------------------------
test_sentences = [
    "I love language models",
    "machine learning models"
]

# -----------------------------
# Tokenization
# -----------------------------
train_tokens = [sent.lower().split() for sent in train_sentences]
test_tokens = [sent.lower().split() for sent in test_sentences]

# -----------------------------
# Build Vocabulary
# -----------------------------
unigram_counts = Counter()
for sent in train_tokens:
    for word in sent:
        unigram_counts[word] += 1

vocab = set(unigram_counts.keys())
word2idx = {word: idx for idx, word in enumerate(vocab)}
idx2word = {idx: word for word, idx in word2idx.items()}
vocab_size = len(vocab)

# -----------------------------
# Prepare Training Data (bigrams)
# -----------------------------
train_data = []
for sent in train_tokens:
    for i in range(len(sent)-1):
        w1, w2 = sent[i], sent[i+1]
        train_data.append((word2idx[w1], word2idx[w2]))

# -----------------------------
# Neural Bigram Language Model
# -----------------------------
class NeuralBigramLM(nn.Module):
    def __init__(self, vocab_size, embedding_dim=50):
        super(NeuralBigramLM, self).__init__()
        self.embeddings = nn.Embedding(vocab_size, embedding_dim)
        self.linear = nn.Linear(embedding_dim, vocab_size)

    def forward(self, context_word_idx):
        embed = self.embeddings(context_word_idx)   # (batch, embed_dim)
        out = self.linear(embed)                   # (batch, vocab_size)
        return out

# -----------------------------
# Training
# -----------------------------
embedding_dim = 50
model = NeuralBigramLM(vocab_size, embedding_dim)
loss_fn = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.01)

for epoch in range(50):
    total_loss = 0
    for w1, w2 in train_data:
        context = torch.tensor([w1], dtype=torch.long)
        target = torch.tensor([w2], dtype=torch.long)

        optimizer.zero_grad()
        output = model(context)
        loss = loss_fn(output, target)
        loss.backward()
        optimizer.step()

        total_loss += loss.item()
    if epoch % 10 == 0:
        print(f"Epoch {epoch}, Loss: {total_loss:.4f}")

# -----------------------------
# Perplexity Evaluation
# -----------------------------
def perplexity(test_tokens, model):
    log_prob = 0
    N = 0
    with torch.no_grad():
        for sent in test_tokens:
            for i in range(len(sent)-1):
                w1, w2 = sent[i], sent[i+1]
                if w1 not in word2idx or w2 not in word2idx:
                    continue
                context = torch.tensor([word2idx[w1]], dtype=torch.long)
                output = model(context)
                prob = torch.softmax(output, dim=1)[0][word2idx[w2]].item()
                log_prob += math.log(prob)
                N += 1
    return math.exp(-log_prob / N)

print("Neural Perplexity:", perplexity(test_tokens, model))
```
