```python
# 1a tokenization by word and sentence
from collections import Counter
from nltk.util import ngrams
import nltk
nltk.download('punkt')

# Given corpus
corpus = [
    "this is my cat",
    "this is my dog",
    "this is your cat",
    "my cat is cute"
]

# Tokenization
tokens = []
for sentence in corpus:
    tokens.extend(sentence.split())

# Unigrams, Bigrams, Trigrams
unigrams = Counter(tokens)
bigrams = Counter(ngrams(tokens, 2))
trigrams = Counter(ngrams(tokens, 3))

# Display results
print("Unigrams:")
print(unigrams)
print("\nBigrams:")
print(bigrams)
print("\nTrigrams:")
print(trigrams)

# Probability of sentence: "this is my cat"
sentence = ["this", "is", "my", "cat"]
probability = 1
V = len(unigrams)  # Vocabulary size

for i in range(len(sentence) - 1):
    bigram = (sentence[i], sentence[i+1])
    probability *= (bigrams[bigram] + 1) / (unigrams[sentence[i]] + V)

print("\nProbability of sentence 'this is my cat' (Bigram Model with Laplace smoothing):")
print(probability)
```

```python
#  4a. tf-IDF
import math
from collections import Counter

# Sample documents
documents = [
    "this is a sample document",
    "this document is another example document",
    "one more sample example"
]

# Tokenize documents
tokenized_docs = [doc.split() for doc in documents]

# Function to calculate TF
def term_frequency(word, document):
    return document.count(word) / len(document)

# Function to calculate IDF
def inverse_document_frequency(word, documents):
    N = len(documents)
    containing_docs = 0

    for doc in documents:
        if word in doc:
            containing_docs += 1

    return math.log(N / containing_docs)

# Vocabulary
vocabulary = set(word for doc in tokenized_docs for word in doc)

print("Term Frequency (TF):\n")

for doc in tokenized_docs:
    print("Document:", doc)
    for word in vocabulary:
        tf = term_frequency(word, doc)
        if tf > 0:
            print(word, "=", round(tf,3))
    print()

print("\nInverse Document Frequency (IDF):\n")

for word in vocabulary:
    idf = inverse_document_frequency(word, tokenized_docs)
    print(word, "=", round(idf,3))
```

```python
# 4b. CYK Parsing Algorithm

# Grammar in CNF form
grammar = {
    'S': ['AB', 'BC'],
    'A': ['BA', 'a'],
    'B': ['CC', 'b'],
    'C': ['AB', 'a']
}

# Input string
string = "baaba"

n = len(string)

# Create empty table
table = [[set() for j in range(n)] for i in range(n)]

# Step 1: Fill first row
for i in range(n):
    for variable in grammar:
        if string[i] in grammar[variable]:
            table[i][i].add(variable)

# Step 2: Fill remaining table
for l in range(2, n+1):       # Length of substring
    for i in range(n-l+1):
        j = i + l - 1
        for k in range(i, j):
            for variable in grammar:
                for rule in grammar[variable]:
                    if len(rule) == 2:
                        if rule[0] in table[i][k] and rule[1] in table[k+1][j]:
                            table[i][j].add(variable)

# Print CYK Table
print("CYK Parsing Table:\n")

for row in table:
    print(row)

# Check acceptance
if 'S' in table[0][n-1]:
    print("\nString is accepted by grammar")
else:
    print("\nString is rejected by grammar")
```

```python
# 5b. probaility of the given statement
# Example Corpus
corpus = [
    "this is my cat",
    "this is your cat",
    "this is a dog",
    "my cat is cute"
]

# Given sentence
sentence = "this is my cat"

# Tokenize corpus
tokenized_corpus = [doc.split() for doc in corpus]

# Tokenize sentence
words = sentence.split()

# Count unigrams
unigram_counts = {}
for doc in tokenized_corpus:
    for word in doc:
        unigram_counts[word] = unigram_counts.get(word, 0) + 1

# Count bigrams
bigram_counts = {}
for doc in tokenized_corpus:
    for i in range(len(doc)-1):
        pair = (doc[i], doc[i+1])
        bigram_counts[pair] = bigram_counts.get(pair, 0) + 1

# Calculate probability
probability = 1

for i in range(len(words)-1):
    w1 = words[i]
    w2 = words[i+1]

    bigram = (w1, w2)

    bigram_count = bigram_counts.get(bigram, 0)
    unigram_count = unigram_counts.get(w1, 1)

    probability *= bigram_count / unigram_count

print("Sentence:", sentence)
print("Probability:", probability)
```

```python
import nltk
from nltk.corpus import gutenberg, stopwords
from nltk.tokenize import word_tokenize
import string
from collections import defaultdict
import math

# Ensure necessary NLTK data is downloaded
print("Downloading 'gutenberg' and 'stopwords' corpora if not already present...")
nltk.download('gutenberg', quiet=True)
nltk.download('stopwords', quiet=True)
print("Download complete.")

# 1. Get a list of file IDs from the NLTK Gutenberg corpus
file_ids = gutenberg.fileids()
print(f"\nTotal number of documents in Gutenberg corpus: {len(file_ids)}")

# 2. Initialize an empty dictionary, document_frequency
document_frequency = defaultdict(int)

# 3. Initialize an empty list, corpus_documents (to store preprocessed words for each doc)
corpus_documents = []

# Get English stopwords for filtering
english_stopwords = set(stopwords.words('english'))

print("\nPreprocessing documents and calculating document frequencies...")
# 4. Loop through each file ID
for fileid in file_ids:
    # a. Load the raw text
    raw_text = gutenberg.raw(fileid)

    # b. Tokenize the document
    words = word_tokenize(raw_text)

    # c. Convert all tokens to lowercase
    words_lower = [word.lower() for word in words]

    # d. Remove punctuation (keeping only alphanumeric words)
    preprocessed_words_doc = [word for word in words_lower if word.isalnum()]

    # e. Filter out English stopwords
    meaningful_words_doc = [word for word in preprocessed_words_doc if word not in english_stopwords]

    # f. Add the unique meaningful words of the current document to a set
    unique_meaningful_words = set(meaningful_words_doc)

    # g. For each word in this document's unique meaningful words set, increment its count
    for word in unique_meaningful_words:
        document_frequency[word] += 1

    # h. Append the list of meaningful words for the current document to corpus_documents
    corpus_documents.append(meaningful_words_doc)

print("Document preprocessing and frequency calculation complete.")

# 5. Determine the total number of documents processed, N
N = len(file_ids)
print(f"Total documents for IDF calculation (N): {N}")

# 6. Define a function calculate_idf
def calculate_idf(word, document_frequency, N):
    # Add a small constant to prevent division by zero if a word exists in no documents (though unlikely here)
    # and to handle log(0) which is undefined, ensuring stability.
    # Using log base e (natural logarithm) as is common in TF-IDF.
    df_word = document_frequency.get(word, 0)
    if df_word == 0:
        return 0.0 # Or some other handling like a very high IDF for unseen words
    return math.log(N / df_word)

# 7. Calculate the IDF for each unique meaningful word
print("\nCalculating IDF scores...")
idf_scores = {}
for word in document_frequency:
    idf_scores[word] = calculate_idf(word, document_frequency, N)
print(f"Calculated IDF scores for {len(idf_scores)} unique words.")

# 8. Print the IDF values for a few illustrative words
print("\nIllustrative IDF values:")
illustrative_words = ['emma', 'book', 'time', 'apple', 'love', 'the', 'is', 'a']
for word in illustrative_words:
    if word in idf_scores:
        print(f"  '{word}': {idf_scores[word]:.6f} (Appears in {document_frequency[word]} documents)")
    else:
        print(f"  '{word}': Not found in corpus (0 documents)")

print("\nExplanation of IDF differences:")
print("  - Words like 'the', 'is', 'a' are very common stopwords and appear in almost all documents, resulting in very low IDF scores (closer to 0). Their presence doesn't help much in distinguishing documents.")
print("  - Words like 'emma' (from 'austen-emma.txt') or 'book' are less common and appear in fewer documents, thus having higher IDF scores. These words are more specific and indicative of a document's content.")
print("  - Words like 'apple' (if not present in the chosen Gutenberg texts) would have no IDF score (or a very high imputed one if smoothed), indicating they are unique or non-existent in this particular corpus.")
```

```python
#9 language model
import math
from collections import defaultdict, Counter

# -----------------------------
# Training Corpus
# -----------------------------
train_sentences = [
    "I love natural language processing",
    "I love machine learning",
    "language models are useful",
    "machine learning is powerful"
]

# -----------------------------
# Test Corpus
# -----------------------------
test_sentences = [
    "I love language models",
    "machine learning models"
]

# Tokenization
train_tokens = [sent.lower().split() for sent in train_sentences]
test_tokens = [sent.lower().split() for sent in test_sentences]

# -----------------------------
# Train Bigram Language Model
# -----------------------------
bigram_counts = defaultdict(Counter)
unigram_counts = Counter()

for sent in train_tokens:
    for i in range(len(sent)-1):
        unigram_counts[sent[i]] += 1
        bigram_counts[sent[i]][sent[i+1]] += 1
    unigram_counts[sent[-1]] += 1

vocab = set(unigram_counts.keys())

# -----------------------------
# Coverage Rate Calculation
# -----------------------------
total_words = 0
covered_words = 0

for sent in test_tokens:
    for word in sent:
        total_words += 1
        if word in vocab:
            covered_words += 1

coverage_rate = covered_words / total_words

print("Coverage Rate:", coverage_rate)

# -----------------------------
# Perplexity Calculation
# -----------------------------
log_prob = 0
N = 0

for sent in test_tokens:
    for i in range(len(sent)-1):
        w1 = sent[i]
        w2 = sent[i+1]
        N += 1

        # Laplace smoothing
        bigram_count = bigram_counts[w1][w2] + 1
        unigram_count = unigram_counts[w1] + len(vocab)

        prob = bigram_count / unigram_count
        log_prob += math.log(prob)

perplexity = math.exp(-log_prob / N)

print("Perplexity:", perplexity)
```

```python
# Import required libraries
import nltk
nltk.download('punkt_tab')
import numpy as np
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from collections import Counter
from sklearn.metrics.pairwise import cosine_similarity

# Download required NLTK resources
nltk.download('punkt')
nltk.download('stopwords')

# Input text
text = input("Enter a paragraph: ")

# Step 1: Sentence Tokenization
sentences = sent_tokenize(text)

# Step 2: Preprocess sentences (remove stopwords)
stop_words = set(stopwords.words('english'))
processed_sentences = []

for sentence in sentences:
    words = word_tokenize(sentence.lower())
    words = [w for w in words if w.isalnum() and w not in stop_words]
    processed_sentences.append(words)

# Step 3: Build vocabulary
vocab = set()
for sentence in processed_sentences:
    vocab.update(sentence)

vocab = list(vocab)

# Step 4: Convert sentences to vectors (Bag of Words)
sentence_vectors = []

for sentence in processed_sentences:
    word_count = Counter(sentence)
    vector = [word_count[word] for word in vocab]
    sentence_vectors.append(vector)

# Step 5: Calculate similarity between sentences
similarity_matrix = cosine_similarity(sentence_vectors)

print("\nSentence Similarity Matrix:")
print(similarity_matrix)

# Step 6: Calculate average similarity (coherence score)
n = len(similarity_matrix)
total = 0
count = 0

for i in range(n):
    for j in range(i+1, n):
        total += similarity_matrix[i][j]
        count += 1

coherence_score = total / count if count > 0 else 0

print("\nCoherence Score:", coherence_score)

# Interpretation
if coherence_score > 0.5:
    print("Text Coherence: High")
elif coherence_score > 0.2:
    print("Text Coherence: Moderate")
else:
    print("Text Coherence: Low")
```

```python
# Import required libraries
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

# Download required NLTK data
nltk.download('punkt')
nltk.download('stopwords')

# Input text
text = input("Enter a paragraph: ")

# Tokenize words
words = word_tokenize(text.lower())

# List of conjunctions and reference words (pronouns)
conjunctions = ['and', 'but', 'or', 'because', 'so', 'although', 'however', 'therefore']
references = ['he', 'she', 'it', 'they', 'them', 'his', 'her', 'their', 'this', 'that', 'these', 'those']

# Stopwords
stop_words = set(stopwords.words('english'))

# Identify cohesive devices
found_conjunctions = [w for w in words if w in conjunctions]
found_references = [w for w in words if w in references]

# Lexical links (repeated meaningful words)
content_words = [w for w in words if w.isalpha() and w not in stop_words]
lexical_links = {}

for w in content_words:
    lexical_links[w] = lexical_links.get(w, 0) + 1

repeated_words = {k: v for k, v in lexical_links.items() if v > 1}

# Display results
print("\nConjunctions found:", found_conjunctions)
print("References (pronouns) found:", found_references)
print("Lexical links (repeated words):", repeated_words)
```

```python
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Embedding, Flatten
from tensorflow.keras.utils import to_categorical

# Sample text data
text = """
machine learning is fun
machine learning is powerful
deep learning is part of machine learning
neural networks are useful in deep learning
"""

# Tokenization
tokenizer = Tokenizer()
tokenizer.fit_on_texts([text])
total_words = len(tokenizer.word_index) + 1

# Create input sequences
input_sequences = []
for line in text.split("\n"):
    token_list = tokenizer.texts_to_sequences([line])[0]
    for i in range(1, len(token_list)):
        n_gram_seq = token_list[:i+1]
        input_sequences.append(n_gram_seq)

# Padding sequences
max_seq_len = max([len(seq) for seq in input_sequences])
input_sequences = np.array(pad_sequences(input_sequences, maxlen=max_seq_len, padding='pre'))

# Split into predictors and label
X = input_sequences[:, :-1]
y = input_sequences[:, -1]

# One hot encoding
y = to_categorical(y, num_classes=total_words)

# Build Feed Forward Neural Network
model = Sequential()
model.add(Embedding(total_words, 10, input_length=max_seq_len-1))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dense(total_words, activation='softmax'))

# Compile model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train model
model.fit(X, y, epochs=200, verbose=1)

# Function to predict next word
def predict_next_word(seed_text):
    token_list = tokenizer.texts_to_sequences([seed_text])[0]
    token_list = pad_sequences([token_list], maxlen=max_seq_len-1, padding='pre')

    predicted = np.argmax(model.predict(token_list), axis=-1)

    for word, index in tokenizer.word_index.items():
        if index == predicted:
            return word

# Test prediction
seed_text = "machine learning"
print("Next word prediction:", predict_next_word(seed_text))
```

```python
#3a.  For chunking using nltk

import nltk
nltk.download('averaged_perceptron_tagger_eng')
from nltk.tokenize import word_tokenize
from nltk import pos_tag, RegexpParser

# Download required resources (run once)
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger')

# Sample sentence
sentence = "Ishita is learning Natural Language Processing at GEC College."

# Tokenize words
words = word_tokenize(sentence)

# POS tagging
pos_tags = pos_tag(words)

# Define chunk grammar (Noun Phrase)
grammar = "NP: {<DT>?<JJ>*<NN.*>+}"

# Create chunk parser
chunk_parser = RegexpParser(grammar)

# Perform chunking
chunk_tree = chunk_parser.parse(pos_tags)

print("Chunking Result:")
print(chunk_tree)

# Draw tree (if running locally)
# chunk_tree.draw()
```

```python
#3b.  Program to perform Named entity Recognition using nltk

import nltk
nltk.download('maxent_ne_chunker_tab')
from nltk.tokenize import word_tokenize
from nltk import pos_tag, ne_chunk

# Download required resources (run once)
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
nltk.download('words')

# Sample text
sentence = "Ishita Singh studies at GEC College in Bhubaneswar, India."

# Tokenize
words = word_tokenize(sentence)

# POS tagging
pos_tags = pos_tag(words)

# Perform Named Entity Recognition
ner_tree = ne_chunk(pos_tags)

print("Named Entity Recognition Result:")
print(ner_tree)

# Draw tree (optional, local system only)
# ner_tree.draw()
```
