```python
from nltk.stem import PorterStemmer

ps = PorterStemmer()
words = ["running", "runs", "ran", "easily"]

for w in words:
    print(w, "→", ps.stem(w))
```

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
words = ["running", "better", "cars"]

for w in words:
    print(w, "→", lemmatizer.lemmatize(w))
```

```python
import nltk
nltk.download('averaged_perceptron_tagger_eng')
from nltk import pos_tag
from nltk.tokenize import word_tokenize

text = "NLTK simplifies natural language processing"
tokens = word_tokenize(text)

pos_tags = pos_tag(tokens)
print(pos_tags)
```

```python
import nltk
nltk.download('maxent_ne_chunker_tab')
nltk.download('words')
from nltk import ne_chunk
from nltk.tokenize import word_tokenize
from nltk import pos_tag

text = "Ravi works at Google in India"
tokens = word_tokenize(text)
tags = pos_tag(tokens)
ner_tree = ne_chunk(tags)

print(ner_tree)
```

```python
from nltk import FreqDist
from nltk.tokenize import word_tokenize

text = "NLP uses machine learning and NLP uses deep learning"
tokens = word_tokenize(text)

fdist = FreqDist(tokens)
print(fdist.most_common(5))
```

```python
from nltk.corpus import wordnet

synonyms = []
antonyms = []

for syn in wordnet.synsets("good"):
    for lemma in syn.lemmas():
        synonyms.append(lemma.name())
        if lemma.antonyms():
            antonyms.append(lemma.antonyms()[0].name())

print("Synonyms:", set(synonyms))
print("Antonyms:", set(antonyms))
```

```python
from nltk.classify import NaiveBayesClassifier

data = [
    ({"good": True}, "positive"),
    ({"bad": True}, "negative")
]

classifier = NaiveBayesClassifier.train(data)
print(classifier.classify({"good": True}))
```
